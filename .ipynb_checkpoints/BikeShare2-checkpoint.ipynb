```python
# Calculate distance between stations
def haversine(p1, p2):
    """
    Calculate the great circle distance b/t two pts
    on earth
    """
    lon1, lat1, lon2, lat2 = map(np.radians, [p1[0],p1[1],p2[0],p2[1]])
    
    # haversine formula
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = np.sin(dlat/2)**2 + np.cos(lat1)*np.cos(lat2)*np.sin(dlon/2)**2
    c = 2*np.arcsin(np.sqrt(a))
    km = 6367*c
    return km
```

```python
# Calculate distance matrix
distances = np.zeros((len(df_stations),len(df_stations)))

for i, row1 in enumerate(df_stations.iterrows()):
    for j, row2 in enumerate(df_stations.iterrows()):
        if i > j:
            continue
        
        distances[i, j] = haversine((row1[1][2],row1[1][0]),
                                    (row2[1][2],row2[1][0]))
        distances[j, i] = distances[i, j]
```

```python
# Plot Distance matrix
fig, ax = plt.subplots(figsize=(15,15))
im = ax.imshow(distances)

ax.set_xticks(np.arange(len(df_stations)))
ax.set_yticks(np.arange(len(df_stations)))

ax.set_xticklabels(df_stations['station_id'])
ax.set_yticklabels(df_stations['station_id'])

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
        rotation_mode="anchor")

# for i in range(len(df_stations)):
#     for j in range(len(df_stations)):
#         text = ax.text(j, i, distances[i, j],
#                       ha="center", va="center", color="w")
        
ax.set_title("Distance Matrix", fontsize=20)
fig.tight_layout()
plt.show()
```

#### For the Duration of the Survey

```python
flow = df_trips[['start_station_id', 'end_station_id']]
flow = flow.dropna()
flow = [tuple(x) for x in flow.to_numpy()]
flow = Counter(list(flow))
flow = [(int(x[0]), int(x[1]), y) for x,y in flow.items()]
```

```python
# Create Graph
G = nx.DiGraph()
G.add_nodes_from(list(df_stations['station_id']))
G.add_weighted_edges_from(flow)
```

```python
net_degree = []
for elem in G.nodes:
    net_degree.append((elem, {
                           "size": G.in_degree(elem, weight="weight") - G.out_degree(elem, weight="weight")
    }))    

scaled_net_degree = []
max_size = max([x[1]["size"] for x in net_degree])
min_size = min([x[1]["size"] for x in net_degree])

for elem, vals in net_degree:
    scaled_net_degree.append((elem, {
       "size": (vals["size"]-min_size) / (max_size-min_size) 
    }))
```

```python
G_scaled = nx.Graph()
G_scaled.add_nodes_from(scaled_net_degree)

d = dict(G_scaled.nodes(data=True)).values()

color_map = {
    "active": "blue",
    "closed": "red",
    "moved": "orange",
    "ACL only": "grey"
}

y_off = -20e-6
x_off = -40e-6
pos = {}
max_lat = max(abs(df_stations['latitude']))
max_long = max(abs(df_stations['longitude']))
colors = []

f, ax = plt.subplots(figsize=(12,12))

for i, row in enumerate(df_stations.iterrows()):
    pos[row[1][4]] = (row[1][2], row[1][0])
    colors.append(color_map[row[1][5]])
#     ax.annotate(row[1][4], 
#                 (pos[row[1][4]][0]+x_off, pos[row[1][4]][1]+y_off))

for label, color in color_map.items():
    ax.plot([0], [0], color=color, label=label)
    
nx.draw_networkx(G_scaled, pos, ax=ax, with_labels=False,
                 node_size=[v["size"]*200 for v in d], node_color=colors)
plt.legend()
ax.set_ylim([30.245, 30.290])
ax.set_xlim([-97.78, -97.70])

plt.axis('off')
plt.show()
```

#### By Month Analysis

```python
time_flow = df_trips[['start_station_id', 'end_station_id', 'month', 'year']]
time_flow = flow.dropna()
time_flow['date'] = flow['month'].astype(str)+ '_' +flow['year'].astype(str)
```

```python
mat = np.zeros((
    42, # number of months in study
    len(df_stations['station_id'])))

station_map = dict((key, i) for i, key in enumerate(list(df_stations['station_id'].value_counts().keys())))

start = 2013*12+12

year = 2013
month = 12

for i, s in enumerate(time_flow.iterrows()):
    month, year = s[1][4].split('_')
    month, year = int(float(month)), int(float(year))
    
    row = year*12+month-start
    
    start_st = station_map[s[1][0]]
    end_st = station_map[s[1][1]]
    
#     print(month)
#     print(year)
    mat[row, start_st] -= 1
    mat[row, end_st] += 1
    
```

```python
# Plot Distance matrix
fig, ax = plt.subplots(figsize=(15,15))
im = ax.imshow(np.transpose(mat))

cbar = ax.figure.colorbar(im, ax=ax)
cbar.ax.set_ylabel("In Degree - Out Degree", fontsize=20, rotation=-90, va="bottom")

ax.set_xlabel("Time", fontsize=16)
ax.set_ylabel("Station ID", fontsize=16)

ax.set_yticks(np.arange(len(df_stations['station_id'])))
ax.set_xticks(np.arange(42))

ax.set_yticklabels(df_stations['station_id'])
ax.set_xticklabels(list(range(42)))

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
        rotation_mode="anchor")

# for i in range(len(df_stations)):
#     for j in range(len(df_stations)):
#         text = ax.text(j, i, distances[i, j],
#                       ha="center", va="center", color="w")


# ax.set_title("Degree By Year", fontsize=20)
fig.tight_layout()
plt.show()
```

```python
# Add graph animation
class AnimateGraph:
    def __init__(self, nodes, mat, pos, colors):
        # Create Graph
        self.G = nx.Graph()
        self.G.add_nodes_from(nodes)
        self.mat = mat
        self.pos = pos
        self.colors = colors
    
    def _update_graph(num):
        ax.clear()
        size = self.mat[:,num]
        nx.draw_networkz(self.G, pos=self.pos, 
                         nod_color=self.color, node_size=size)

    def animate_graph():
        fig, ax = plt.subplots(figsize=(12, 8))
        plt.axis('off')
        
        # Create animation
        ani = animation.FuncAnimation(fig,self._update_graph,
                                     frames=mat.shape[1])
        ani.save('anim.gif', writer-'imagemagick')
        plt.close(ani._fig)
    
```

```python
animate = False
display_animation = False

if animate:
    nodes = list(df_stations['station_id'])
    animate = AnimateGraph(nodes, mat, pos, colors)
    animate.animate_graph()
    
if display_animation:
    with open('anim.gif', 'rb') as file:
        display(Image(file.read()))
```
